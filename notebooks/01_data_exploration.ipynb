### Next Milestones (4.30 - 4.35)

Now that the data is loaded, let's inspect the shapes, column types, fetch basic statistics, find missing values, and check for duplicates.

```python
# Milestone 4.30: Basic summary statistics for individual columns
df.describe()
```

```python
# Milestone 4.31, 4.32: Shapes and slicing
print("Data Shape:", df.shape)

# Let's slice the first two columns of the first 3 rows as an example
print("\nSlicing (Rows 0-2, Columns 0-2):")
df.iloc[:3, :2]
```

```python
# Milestone 4.33, 4.34: Detecting and Handling missing values
print("Missing values per column:\\n", df.isnull().sum())

# Example drop missing values strategy:
# df_clean = df.dropna() 

# Example fill strategy:
# df_filled = df.fillna(0)
```

```python
# Milestone 4.35: Removing duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# df_no_duplicates = df.drop_duplicates()
```

```python
# Milestone 4.36: Standardizing Column Names and Data Formats
# Let's make sure all column names are lowercase and have no spaces
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Drop any missing values and duplicate rows to create a clean dataset
df_clean = df.dropna().drop_duplicates()

print(f"Original shape: {df.shape}")
print(f"Clean shape: {df_clean.shape}")

# Save the clean dataset to our processed data folder
df_clean.to_csv('../data/processed/teaching_quality_clean.csv', index=False)
print("\\nData successfully cleaned and saved to data/processed/teaching_quality_clean.csv")
```
