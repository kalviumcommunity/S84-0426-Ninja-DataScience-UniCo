# Data Visualization & Exploratory Data Analysis

**Milestones 4.37 - 4.43:** Extracting actionable insights on teaching quality and student engagement using visual tools.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean style for our dashboard-like charts
sns.set_theme(style="whitegrid")

# Load the clean dataset from Phase 1
df = pd.read_csv('../data/processed/teaching_quality_clean.csv')
df.head()
```

### 1. Visualizing Data Distributions (Milestone 4.39)
Let's look at the distribution of teaching feedback scores using a **Histogram**. This will immediately tell us if most teachers are rated highly or poorly.

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='feedback_score', bins=10, kde=True, color='royalblue')
plt.title('Distribution of Teaching Feedback Scores', fontsize=14)
plt.xlabel('Feedback Score (e.g., 1 to 5)')
plt.ylabel('Frequency')
plt.show()
```

### 2. Detecting Outliers using Boxplots (Milestones 4.40 & 4.43)
Are there any extreme low or high attendance rates? A **Boxplot** is great for quickly spotting outliers and understanding the spread of student engagement.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='attendance_rate', color='orange')
plt.title('Boxplot for Attendance Rates (Outlier Detection)', fontsize=14)
plt.xlabel('Attendance Rate (%)')
plt.show()
```

### 3. Exploring Relationships (Milestone 4.42)
**Core Problem Statement Goal:** Does higher attendance correlate with better feedback? Let's use a **Scatter Plot** to find out.

```python
plt.figure(figsize=(8, 5))
# Note: If 'course_id' or 'attendance_rate' differ slightly in your CSV, we can adjust the column names.
if 'attendance_rate' in df.columns and 'feedback_score' in df.columns:
    sns.scatterplot(data=df, x='attendance_rate', y='feedback_score', alpha=0.7, color='teal')
    plt.title('Relationship: Attendance Rate vs Feedback Score', fontsize=14)
    plt.xlabel('Attendance Rate (%)')
    plt.ylabel('Feedback Score')
    plt.show()
else:
    print("Verify column names in df.head() above to generate this scatterplot.")
```
